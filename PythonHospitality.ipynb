```python
plt.figure(figsize=(8, 4))
ax1 = sns.countplot(x='hotel', hue='is_canceled', data=df, palette='Blues')
legend_labels, _ = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor(1,1))
plt.title('Reservation status in different hotels', size=20)
plt.xlabel('hotel')
plt.ylabel('number of reservations')
plt.show
```

```python

resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20, 8))
plt.title('Average Daily Rate in City and Resort Hotel', fontsize=30)  # Corrected the syntax for setting the title
plt.plot(resort_hotel.index, resort_hotel['adr'], label='Resort_Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label='City_Hotel')
plt.legend(fontsize=20)
plt.show()

```

```python
df['month'] = df['reservation_status_date'].dt.month  # Added missing dot and corrected dt.month
plt.figure(figsize=(16, 8))  # Added closing parenthesis and corrected indentation
ax1 = sns.countplot(x='month', hue='is_canceled', data=df, palette='bright')  # Corrected countplot syntax and palette name
legend_labels, _ = ax1.get_legend_handles_labels()  # Corrected method name
ax1.legend(bbox_to_anchor=(1, 1))
plt.title('Reservation status per month', size=20)  # Added missing comma
plt.xlabel('month')
plt.ylabel('number of reservations')
plt.legend(['not canceled', 'canceled'])
plt.show()

```
